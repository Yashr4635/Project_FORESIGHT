```python
import pandas as pd
```

```python
df = pd.read_csv("../data/processed/feature_engineered.csv")

df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])

df.head()
```

```python
monthly_revenue = (
    df.groupby(df["InvoiceDate"].dt.to_period("M"))["Revenue"]
    .sum()
    .reset_index()
)

monthly_revenue["InvoiceDate"] = monthly_revenue["InvoiceDate"].astype(str)

monthly_revenue.head()
```

```python
top_products = (
    df.groupby("Description")["Revenue"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

top_products
```

```python
country_sales = (
    df.groupby("Country")["Revenue"]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

country_sales.head()
```

```python
customer_sales = (
    df.groupby("Customer ID")["Revenue"]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

customer_sales.head()
```

```python
monthly_revenue.to_csv(
    "../data/processed/dashboard_monthly.csv",
    index=False
)

top_products.to_csv(
    "../data/processed/dashboard_products.csv",
    index=False
)

country_sales.to_csv(
    "../data/processed/dashboard_country.csv",
    index=False
)

customer_sales.to_csv(
    "../data/processed/dashboard_customers.csv",
    index=False
)

print("Dashboard datasets created successfully!")
```
