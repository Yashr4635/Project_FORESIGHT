```python
import pandas as pd

# Load cleaned dataset
df = pd.read_csv("../data/processed/cleaned_step1.csv")

# Create Revenue column
df["Revenue"] = df["Quantity"] * df["Price"]

# Convert InvoiceDate to datetime
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])

df.head()
```

```python
df["Year"] = df["InvoiceDate"].dt.year

df["Year"].head()
```

```python
df["Month"] = df["InvoiceDate"].dt.month

df["Month"].head()
```

```python
df["Day"] = df["InvoiceDate"].dt.day

df["Day"].head()
```

```python
df["Weekday"] = df["InvoiceDate"].dt.day_name()

df["Weekday"].head()
```

```python
df["Hour"] = df["InvoiceDate"].dt.hour

df["Hour"].head()
```

```python
df["IsWeekend"] = df["Weekday"].isin(["Saturday", "Sunday"])

df.head()
```

```python
basket_value = (
    df.groupby("Invoice")["Revenue"]
      .sum()
)

basket_value.head()
```

```python
customer_spending = (
    df.groupby("Customer ID")["Revenue"]
      .sum()
)

customer_spending.head()
```

```python
purchase_frequency = (
    df.groupby("Customer ID")["Invoice"]
      .nunique()
)

purchase_frequency.head()
```

```python
avg_basket = (
    df.groupby("Invoice")["Quantity"]
      .sum()
)

avg_basket.describe()
```

```python
df.to_csv(
    "../data/processed/feature_engineered.csv",
    index=False
)

print("Feature Engineering Completed!")
```
